# NULL VALUES

```python
data.fillna("No description for this channel",inplace=True)
```

```python
def convert_subscribers(subscribers):
    if 'M' in subscribers:
        return float(subscribers.replace('M', '')) * 1e6  # Convert to millions
    else:
        return float(subscribers)

# Apply the function to the 'Subscribers' column
data['Subscribers'] = data['Subscribers'].apply(convert_subscribers)

# Change the data type of 'Subscribers' to int or float
data['Subscribers'] = data['Subscribers'].astype(int)  # Change to int
```

```python
def convert_views(views):
    if 'B' in views:
        return float(views.replace('B', '')) * 1e9  # Convert to millions
    else:
        return float(views)

# Apply the function to the 'Subscribers' column
data['Views'] = data['Views'].apply(convert_views)

# Change the data type of 'Subscribers' to int or float
data['Views'] = data['Views'].astype(int)  # Change to int
```

# VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_10_channels = data.sort_values(by='Subscribers', ascending=False).head(10)

# Create a bar plot for the top ten Subscribers
plt.figure(figsize=(10, 6))
sns.barplot(x='Channel Name', y='Subscribers', data=top_10_channels)
plt.title('Top 10 Channels by Subscribers')
plt.xticks(rotation=45)
plt.xlabel('Channel Name')
plt.ylabel('Subscribers')
plt.show()
```

```python
# Create a scatter plot for Views vs. Videos
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Views', y='Videos', data=data, hue='Subscribers', palette='viridis')
plt.title('Views vs. Videos')
plt.xlabel('Views')
plt.ylabel('Videos')
plt.legend(title='Subscribers')
plt.show()
```

```python
# Get the top 25 countries by subscriber count
top_25_channels = data['Channel Name'].value_counts().head(25)

# Create a pie chart for the top 25 countries
plt.figure(figsize=(8, 8))
plt.pie(top_25_channels, labels=top_25_channels.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Subscribers by Channels (Top 25)')
plt.show()
```

```python
data1=data.drop(["Channel Name","User Name","Channel Description"],axis=1)
correlation_matrix = data1.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
